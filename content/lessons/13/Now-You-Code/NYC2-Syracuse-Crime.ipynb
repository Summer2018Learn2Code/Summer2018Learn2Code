```python
# Step 1:
# Now plot the same thing using plot.ly without cufflinks, again filling in the templates. Using plot.ly takes more effort but
# you have more control / flexibility over how the plots are rendered.
plot_data = [
    go.Scatter(x=crime['Year'], y=crime['Robbery rate'], mode="lines+markers", name="Robbery Rate", marker=dict(color='orange')),
    go.Scatter(x=crime['Year'], y=crime['Motor vehicle theft rate'], mode="lines+markers", name="Motor Vehicle Theft Rate", marker=dict(color='blue')),
]

py.iplot({ 'data': plot_data,
            'layout': {
               'title': 'Robbery Rate versus Motor Vehicle Theft Rate',
               'xaxis': {
                 'title': 'Year'},
               'yaxis': {
                'title': 'Rate'}
        }})
```
